# 02 Data Cleaning and Preprocessing

This notebook documents the data cleaning procedures applied to the raw mutual fund datasets. The cleaning pipeline handles duplicate values, formats datetime variables, normalizes text columns, and prepares the data for ingestion into the SQLite analytical database.

```python
import pandas as pd
import numpy as np
from pathlib import Path

RAW_DIR = Path("../data/raw")
PROCESSED_DIR = Path("../data/processed")
```

### 1. Cleaning Fund Master Dataset

We clean the fund master dataset by removing duplicates, trimming whitespace from textual columns, and parsing launch dates.

```python
fund_df = pd.read_csv(RAW_DIR / "01_fund_master.csv")
print(f"Original rows: {len(fund_df)}")

# Drop duplicates and clean text
fund_df = fund_df.drop_duplicates()
text_cols = ["fund_house", "scheme_name", "category", "sub_category", "plan", "benchmark", "fund_manager", "risk_category"]
for col in text_cols:
    if col in fund_df.columns:
        fund_df[col] = fund_df[col].astype(str).str.strip()

print(f"Cleaned rows: {len(fund_df)}")
fund_df.head()
```

### 2. Cleaning NAV History Dataset

We parse dates, handle numeric precision, and ensure date order consistency.

```python
nav_df = pd.read_csv(RAW_DIR / "02_nav_history.csv")
print(f"Original NAV rows: {len(nav_df)}")
nav_df["date"] = pd.to_datetime(nav_df["date"])
nav_df = nav_df.sort_values(["amfi_code", "date"])
print(f"Cleaned NAV rows: {len(nav_df)}")
```
